```python
from scipy import optimize
import numpy as np
from matplotlib import pylab as plt
import math
%matplotlib inline

def f(x):
    return 0.1*x**5+x**4-3*x**3+2*x**2-17*x+23

x = np.arange(-10, 10, 0.1)
y = f(x)
plt.plot(x, y)
plt.show()
```

```python
print(optimize.minimize(f, [1], method='Nelder-Mead'))
```

```python
print(optimize.minimize(f, [2], method='BFGS'))
```
