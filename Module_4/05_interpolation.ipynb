```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
def f(x):
    return np.sin(x / 5)*np.exp(x / 10)+ 5*np.exp(-x / 2)
```

```python
x = np.arange(1, 16, 0.1)
y=f(x)
```

```python
plt.plot(x, y, '-', x, y, '-')
plt.show()
```

```python
a1=1.0
a2=15.0
b1=f(a1)
b2=f(a2)
print(a1, b1)
print(a2, b2)
```

```python
a = np.array([[1, a1], [1, a2]])
b = np.array([b1,b2])
x = np.linalg.solve(a, b)
x
```

```python
np.allclose(np.dot(a, x), b)
```

```python
X = np.arange(1, 16, 0.1)
y = x[0] + x[1]*X
```

```python
plt.plot(X, y, '-', X, y, '-')
plt.show()
```

```python
a1=1.0
a2=8.0
a3=15.0
b1=f(a1)
b2=f(a2)
b3=f(a3)
print(a1, b1)
print(a2, b2)
print(a3, b3)
```

```python
a = np.array([[1, a1, a1**2], [1, a2, a2**2],[1, a3, a3**2]])
b = np.array([b1, b2, b3])
x = np.linalg.solve(a, b)
x
```

```python
np.allclose(np.dot(a, x), b)
```

```python
X = np.arange(1, 16, 0.1)
y = x[0]+x[1]*X + x[2]*X**2
```

```python
plt.plot(X, y, '-', X, y, '-')
plt.show()
```
